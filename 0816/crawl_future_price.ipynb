```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
df = pd.read_csv('../0806/tapei_tripadvisor_top300.csv')
all_hotel = df['title'].to_list()
all_uri = df['uri'].to_list()
```

```python
import re
import json
```

```python
def get_price(url_list):
    for url in url_list:
        time.sleep(2)
        resp = requests.get(url)
        soup = BeautifulSoup(resp.text,'html.parser')
        # 找放價格的 js
        Js = soup.select('script[type="text/javascript"]')
        priceJs = soup.findAll(attrs={'type':'text/javascript'}, string=re.compile(r'JSON.parse'))
        for p in priceJs:
            if 'servlets/responsivehotelreview' in p.text:
                try:
                    text = p.text.split('YELLOW_BRICKS_3_1')[1]
#                     print(p.text.split('YELLOW_BRICKS_3_1')[1])
                except:
                    print('error')
    return text
```

```python
price_text = get_price(all_uri[:1])
```
